```python
# Import the necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set() # setting seaborn default for plots

%matplotlib inline

# Initialize Figure and Axes object
fig, ax = plt.subplots()

# Load in data
df = pd.read_csv("WHO-COVID-19-global-data.csv")

# Create violinplot
ax.violinplot(df["New_cases"], vert=False)

# Show the plot
plt.show()
```

```python
# Load the data
df = pd.read_csv("WHO-COVID-19-global-data.csv")

# Create violinplot
sns.violinplot(x = "New_cases", data=df)

# Show the plot
plt.show()
```

```python
# Load the data
df = pd.read_csv("WHO-COVID-19-global-data.csv")

# Create violinplot
sns.violinplot(x = "Cumulative_cases", data=df)

# Show the plot
plt.show()
```

```python
df=pd.read_csv('WHO-COVID-19-global-data.csv')
sns.violinplot(x="New_deaths",data=df)
plt.show()
```

```python
p=sns.countplot(x='New_cases',data=df)
p.set_xlabel('0=New_cases,1=New_deaths')
p.set_title('Survival count plot')
sns.despine()
```

```python
sns.boxplot(x='Cumulative_cases',data=df)
```
